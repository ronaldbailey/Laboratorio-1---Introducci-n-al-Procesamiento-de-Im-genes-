```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def color_filter(image, color):
    # Crear una matriz de ceros del mismo tamaño que la imagen para usar como máscara
    zeros = np.zeros(image.shape[:2], dtype="uint8")

    # Dividir la imagen en sus canales de color
    blue, green, red = cv2.split(image)

    if color == 1:  # Azul
        image = cv2.merge([blue, zeros, zeros])
    elif color == 2:  # Verde
        image = cv2.merge([zeros, green, zeros])
    elif color == 3:  # Rojo
        image = cv2.merge([zeros, zeros, red])
    elif color == 10:  # Rojo y Verde
        image = cv2.merge([zeros, green, red])
    elif color == 20:  # Verde y Azul
        image = cv2.merge([blue, green, zeros])
    elif color == 30:  # Azul y Rojo
        image = cv2.merge([blue, zeros, red])
    else:
        print("El valor del color no es válido. Debe ser 1, 2, 3, 10, 20 o 30.")

    return image
```

# Uso de la funcion

```python


# Cargar la imagen
string = './Lab1/imagen1/imagen1_salida_gray_azul.jpg'
imagen = cv2.imread(string)

# Aplicar el filtro de color
filtered_img = color_filter(imagen, 1)

# Convertir la imagen de BGR a RGB
filtered_img = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(filtered_img)
plt.show()

# Aplicar el filtro de color
filtered_img = color_filter(imagen, 2)

# Convertir la imagen de BGR a RGB
filtered_img = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(filtered_img)
plt.show()

# Aplicar el filtro de color
filtered_img = color_filter(imagen, 3)

# Convertir la imagen de BGR a RGB
filtered_img = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(filtered_img)
plt.show()


# Aplicar el filtro de color
filtered_img = color_filter(imagen, 10)

# Convertir la imagen de BGR a RGB
filtered_img = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(filtered_img)
plt.show()

# Aplicar el filtro de color
filtered_img = color_filter(imagen, 20)

# Convertir la imagen de BGR a RGB
filtered_img = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(filtered_img)
plt.show()

# Aplicar el filtro de color
filtered_img = color_filter(imagen, 30)

# Convertir la imagen de BGR a RGB
filtered_img = cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(filtered_img)
plt.show()



```
