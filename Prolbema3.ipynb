```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def create_grayscale_images(image_path):
    # Leer la imagen original
    image = cv2.imread(image_path)

    # Convertir la imagen de BGR a RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Dividir la imagen en sus canales de color
    red_channel, green_channel, blue_channel = cv2.split(image)

    # Guardar cada canal como una imagen en escala de grises
    cv2.imwrite('./Lab1/out/red_channel.jpg', red_channel)
    cv2.imwrite('./Lab1/out/green_channel.jpg', green_channel)
    cv2.imwrite('./Lab1/out/blue_channel.jpg', blue_channel)

    return 'red_channel.jpg', 'green_channel.jpg', 'blue_channel.jpg'

# Ruta a la imagen original
image_path = './Lab1/odish.jpg'

# Crear las imgenes en escala de grises
red_image_path, green_image_path, blue_image_path = create_grayscale_images(image_path)

# Mostrar las imgenes en escala de grises
red_image = cv2.imread(red_image_path, cv2.IMREAD_GRAYSCALE)
green_image = cv2.imread(green_image_path, cv2.IMREAD_GRAYSCALE)
blue_image = cv2.imread(blue_image_path, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(red_image, cmap='gray')
plt.title('Rojo')

plt.subplot(1, 3, 2)
plt.imshow(green_image, cmap='gray')
plt.title('Verde')

plt.subplot(1, 3, 3)
plt.imshow(blue_image, cmap='gray')
plt.title('Azul')

plt.show()
```
