```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_histograms(image_path):
    # Leer la imagen original
    image = cv2.imread(image_path)

    # Convertir la imagen de BGR a RGB
    image = image[:, :, ::-1]

    # Dividir la imagen en sus canales de color
    red_channel = image[:, :, 0]
    green_channel = image[:, :, 1]
    blue_channel = image[:, :, 2]

    # Calcular la escala de grises como el promedio de los canales de color
    grayscale = np.mean(image, axis=2).astype(np.uint8)

    # Crear los histogramas
    colors = ('r', 'g', 'b', 'gray')
    for i, channel in enumerate([red_channel, green_channel, blue_channel, grayscale]):
        histogram, bins = np.histogram(channel.flatten(), bins=256, range=[0,256])
        plt.plot(histogram, color=colors[i])
        plt.axvline(np.mean(channel), color=colors[i], linestyle='dashed', linewidth=2)
        plt.title(f'{colors[i]} Histogram')
        plt.show()

# Ruta a la imagen original
image_path = './Lab1/odish.jpg'

plt.imshow(cv2.imread(image_path,cv2.COLOR_BGR2RGB))
plt.show()
# Crear los histogramas
plot_histograms(image_path)
plt.show()
```
