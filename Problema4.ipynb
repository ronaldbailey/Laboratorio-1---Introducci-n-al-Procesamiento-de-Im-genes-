```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def trazar_histogramas(ruta_imagen):
    # Leer la imagen original
    imagen = cv2.imread(ruta_imagen)

    # Convertir la imagen de BGR a RGB
    imagen = imagen[:, :, ::-1]

    # Dividir la imagen en sus canales de color
    canal_rojo = imagen[:, :, 0]
    canal_verde = imagen[:, :, 1]
    canal_azul = imagen[:, :, 2]

    # Calcular la escala de grises como el promedio de los canales de color
    escala_grises = np.mean(imagen, axis=2).astype(np.uint8)

    # Crear los histogramas
    colores = ('r', 'g', 'b', 'grey')
    for i, canal in enumerate([canal_rojo, canal_verde, canal_azul, escala_grises]):
        histograma, bins = np.histogram(canal.flatten(), bins=256, range=[0,256])
        plt.plot(histograma, color=colores[i])
        plt.axvline(np.mean(canal), color=colores[i], linestyle='dashed', linewidth=2)
        plt.title(f'Histograma {colores[i]}')
        plt.show()

# Ruta a la imagen original
ruta_imagen = './Lab1/odish.jpg'

plt.imshow(cv2.imread(ruta_imagen,cv2.COLOR_BGR2RGB))
plt.show()
# Crear los histogramas
trazar_histogramas(ruta_imagen)
plt.show()
```
