```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def reconstruct_image(blue_image_path, green_image_path, red_image_path):
    # Leer las im√°genes en escala de grises
    blue_image = cv2.imread(blue_image_path, cv2.IMREAD_GRAYSCALE)
    green_image = cv2.imread(green_image_path, cv2.IMREAD_GRAYSCALE)
    red_image = cv2.imread(red_image_path, cv2.IMREAD_GRAYSCALE)

  
    # Crear una imagen temporal con 3 canales
    temp_image = np.dstack((blue_image, green_image, red_image))

    # Crear una nueva imagen con la misma forma y tipo que la imagen temporal
    reconstructed_image = np.zeros_like(temp_image)

    # Asignar cada canal de color
    reconstructed_image[:,:,0] = blue_image
    reconstructed_image[:,:,1] = green_image
    reconstructed_image[:,:,2] = red_image

    return reconstructed_image

```

# Usar la funcion


```python

# Rutas a las im√°genes en escala de grises
blue_image_path = './Lab1/imagen1/imagen1_salida_gray_azul.jpg'
green_image_path = './Lab1/imagen1/imagen1_salida_gray_verde.jpg'
red_image_path = './Lab1/imagen1/imagen1_salida_gray_rojo.jpg'

# Reconstruir la imagen original
reconstructed_image = reconstruct_image(blue_image_path, green_image_path, red_image_path)

# Convertir la imagen de BGR a RGB
reconstructed_image = cv2.cvtColor(reconstructed_image, cv2.COLOR_BGR2RGB)

# Mostrar la imagen reconstruida
plt.imshow(reconstructed_image)
plt.show()
```
